## Matplotlib is used for generating graphs.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b)
plt.show()
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b,"*")
plt.show()
```

```python
# multiple points Matplotlib dotted
a = np.array([1,2,3,7,9,10])
b = np.array([4,5,6,13,22,14])
plt.plot(a,b, linestyle= "--")
plt.show()
```

```python
# Matplotlib marker
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b, marker="o")
plt.show()
```

```python
# Marker size
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b, marker="o", ms = 5)
plt.show()
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b, marker="o", ms = 5, mec = 'r')
plt.show()
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
plt.plot(a,b, marker="o", ms = 5, mec = 'r', mfc = 'g')
plt.show()
```

# Matplotlib Line

```python
x = np.array([2,4,6])
y = np.array([4,8,10])
plt.plot(x,y, linestyle="--")
plt.show()


```

```python
x = np.array([2,4,6])
y = np.array([4,8,10])
plt.plot(x,y, color = 'r',marker='o',mec ='green',mfc = 'blue', linestyle="--")
plt.show()
```

```python
#Linewidth
x = np.array([2,4,6])
y = np.array([4,8,10])
plt.plot(x,y, color = 'r',marker='o',mec ='green',mfc = 'blue', linestyle="--", linewidth= 3)
plt.show()
```

```python
# Multiple lines
x = np.array([2,4,6])
y = np.array([4,8,10])
plt.plot(x, color = 'r',marker='o',mec ='green',mfc = 'blue', linestyle="--")
plt.plot(y, color = 'b',marker='*',mec ='yellow',mfc = 'gray', linestyle="dotted")
plt.show()
```

```python
# Title
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Fitness Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [7,3,10,11,13]
y2 = [ 2,4,6,8,10]

# Create a plot
plt.plot(x,y1, label ="Quadratic", color = 'blue')
plt.plot(x,y2, label ="Linear", color = 'green')

plt.legend()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Example plot with legend")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = {'family': 'serif', 'color':'blue', 'size':20})
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# Position of the title = left, right, centre and identified as "Loc"
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(axis='y')
plt.show()
```
